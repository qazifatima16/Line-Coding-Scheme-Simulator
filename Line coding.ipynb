```python
import matplotlib.pyplot as plt
#%matplotlib inline
#from array import *
```

```python
coding_scheme=input("Enter the line coding scheme:\n1.NRZ-L\n2.NRZ-I\n3.Manchester\n4.Differential Manchester\n5.AMI\n")
```

```python
if coding_scheme=='5':
    scrambling_choice=input("Is scrambling needed?\n Yes or No\n")
```

```python
if coding_scheme=='5' and (scrambling_choice=='yes' or scrambling_choice=='Yes'):
    scrambling=input("Enter the scrambling type.\n1.B8ZS\n2.HDB3\n")
```

```python
data=input('Enter the data stream\n').split(" ")
```

```python
len_data=len(data)
```

```python
if coding_scheme=="1":
    print(data)
    
    fig=plt.figure()
    ax=fig.add_subplot(111)
    ax.set(title='NRZ-L Signal',xlim=[0,len_data],ylim=[-2,2],ylabel='Amplitude',xlabel='Time Unit')
    
    zero_line=[0]
    v_line_x=[]
    y=[]
    for i in range(len(data)):
        zero_line.append(0)
        v_line_x.append(i+1)
        v_line_x.append(i+1)
        
        if int(data[i])==1:
            y.append(1)
        else:
            y.append(-1)
            
        xx=[0,0]
        yy=[0,0]
        
        xx[0]=i
        xx[1]=int(1)+i
        yy[0]=yy[1]=int(y[i])
        plt.plot(xx,yy,'r')
        plt.plot(zero_line,'c')
        plt.plot(v_line_x,[2,-2],'c--')
            
        v_line_x.pop()
        v_line_x.pop()    
            
        if i!=0 and y[i-1]!=y[i]:
            v_line_x.append(i)
            v_line_x.append(i)
            plt.plot(v_line_x,[1,-1],'r')
            v_line_x.pop()
            v_line_x.pop() 
        
    plt.show()
    
```

```python
if coding_scheme=="2":
    print(data)
    
    fig=plt.figure()
    ax=fig.add_subplot(111)
    ax.set(title='NRZ-I Signal',xlim=[0,len_data],ylim=[-2,2],ylabel='Amplitude',xlabel='Time Unit')
    
    zero_line=[0]
    v_line_x=[]
    y=[]

    for i in range(len(data)):
        zero_line.append(0)
        v_line_x.append(i+1)
        v_line_x.append(i+1)
        
        if i==0:
            if int(data[i])==1:
                y.append(1)
            else:
                y.append(-1)
            
        else:
            if int(data[i])==1:
                if y[i-1]==1:
                    y.append(-1)
                else:
                    y.append(1)
            else:
                y.append(y[i-1])
            
        xx=[0,0]
        yy=[0,0]
        
        xx[0]=i
        xx[1]=int(1)+i
        yy[0]=yy[1]=int(y[i])
        plt.plot(xx,yy,'r')
        plt.plot(zero_line,'c')
        plt.plot(v_line_x,[2,-2],'c--')
            
        v_line_x.pop()
        v_line_x.pop()    
            
        if i!=0 and y[i-1]!=y[i]:
            v_line_x.append(i)
            v_line_x.append(i)
            plt.plot(v_line_x,[1,-1],'r')
            v_line_x.pop()
            v_line_x.pop() 
            
    plt.show()
    
```

```python
if coding_scheme=="3":
    print(data)
    
    fig=plt.figure()
    ax=fig.add_subplot(111)
    ax.set(title='Manchester Signal',xlim=[0,len_data],ylim=[-2,2],ylabel='Amplitude',xlabel='Time Unit')
    
    zero_line=[0]
    v_line_x=[]
    v_line_v=[0,0]
    y=[]
    for i in range(len(data)):
        zero_line.append(0)
        v_line_x.append(i+1)
        v_line_x.append(i+1)
        
        if int(data[i])==1:
            y.append(-1)
            y.append(1)
        else:
            y.append(1)
            y.append(-1)

        xx=[0,0,0,0]
        yy=[0,0,0,0]

        xx[0]=i
        xx[1]=xx[2]=i+0.5
        xx[3]=i+1
        yy[0]=yy[1]=y[i*2]
        yy[2]=yy[3]=y[(i*2)+1]
        plt.plot(xx,yy,'r')
        plt.plot(zero_line,'c')
        v_line_v[0]=i+0.5
        v_line_v[1]=i+0.5
        plt.plot(v_line_x,[2,-2],'c--')
        plt.plot(v_line_v,[1,-1],'r')
            
        if y[i]!=0 and y[i*2-2]==y[i*2]:
            v_line_v[0]=i
            v_line_v[1]=i
            plt.plot(v_line_v,[1,-1],'r')
            
        v_line_x.pop()
        v_line_x.pop()    
            
       
    plt.show()
    
```

```python
if coding_scheme=="4":
    print(data)
    
    fig=plt.figure()
    ax=fig.add_subplot(111)
    ax.set(title='Differential Manchester Signal',xlim=[0,len_data],ylim=[-2,2],ylabel='Amplitude',xlabel='Time Unit')
    
    zero_line=[0]
    v_line_x=[]
    v_line_v=[0,0]
    y=[]
        
    for i in range(len(data)):
        zero_line.append(0)
        v_line_x.append(i+1)
        v_line_x.append(i+1)
        
        if i==0:
            if int(data[0])==1:
                y.append(-1)
                y.append(1)
            else:
                y.append(1)
                y.append(-1)
                
        else:
            if int(data[i])==1:
                y.append((y[len(y)-2])*(-1))
                y.append((y[len(y)-2])*(-1))
            else:
                y.append((y[len(y)-2]))
                y.append((y[len(y)-2]))

        xx=[0,0,0,0]
        yy=[0,0,0,0]

        xx[0]=i
        xx[1]=xx[2]=i+0.5
        xx[3]=i+1
        yy[0]=yy[1]=y[i*2]
        yy[2]=yy[3]=y[(i*2)+1]
        plt.plot(xx,yy,'r')
        plt.plot(zero_line,'c')
        v_line_v[0]=i+0.5
        v_line_v[1]=i+0.5
        plt.plot(v_line_x,[2,-2],'c--')
        plt.plot(v_line_v,[1,-1],'r')
            
        if y[i]!=0 and y[(i*2)-1]!=y[i*2]:
            v_line_v[0]=i
            v_line_v[1]=i
            plt.plot(v_line_v,[1,-1],'r')
            
        v_line_x.pop()
        v_line_x.pop()    
            
       
    plt.show()
```
